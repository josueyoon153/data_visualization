```python

### LIBRERIAS Y LECTURA DE DATOS
import pandas as pd
import numpy as np
import datetime as dt
from IPython.display import display
import ipywidgets as widgets


import plotly as py # interactive graphing
from plotly.offline import init_notebook_mode, iplot, plot
from plotly.graph_objs import *
import plotly.tools as tls

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from plotly.offline import init_notebook_mode, iplot, plot
from plotly.graph_objs import *

import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
RIB = pd.read_csv(r'G:\Mi unidad\GCBA\SECPECG\DGEVC\03_MDH\RIB\RIB_activo_col_usable_ACOTADO.csv')
```

```python
ALL = 'ALL'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique
```

```python
RIB['Nombre plan'] = RIB['Nombre plan'].astype(str)
```

```python
dropdown_planes_2 = widgets.Dropdown(options = unique_sorted_values_plus_ALL(RIB['Nombre plan']))
output = widgets.Output()
plot_output = widgets.Output()



def common_filtering(plan):
    output.clear_output()
    plot_output.clear_output()
    
    if (plan == ALL):
        common_filter = RIB
    else:
        common_filter = RIB[RIB['Nombre plan'] == plan]
        
    with plot_output:
        import plotly.express as px
        common_filter_2 = pd.DataFrame(common_filter.groupby('TIPO_PERSONA').count()['RIB_BENEFICIARIO'])
        common_filter_2.reset_index(inplace=True)
        common_filter_2.sort_values('RIB_BENEFICIARIO', ascending=False, inplace=True)
        fig = px.treemap(common_filter_2, path=['TIPO_PERSONA'], values='RIB_BENEFICIARIO')
        plotly.offline.iplot(fig)

def dropdown_plan_eventhandler2(change):
    common_filtering(change.new)


dropdown_planes_2.observe(dropdown_plan_eventhandler2, names='value')
display(dropdown_planes_2)
display(plot_output)
```

```python
output = widgets.Output()
plot_output = widgets.Output()

dropdown_planes_1 = widgets.Dropdown(options = unique_sorted_values_plus_ALL(RIB['Nombre plan']), description='Plan:')

def common_filtering(plan):
    output.clear_output()
    plot_output.clear_output()
    
    if (plan == ALL):
        common_filter = RIB
    else:
        common_filter = RIB[RIB['Nombre plan'] == plan]
        
    with plot_output:
        import plotly.express as px
        fig = px.box(common_filter, y="EDAD")
        plotly.offline.iplot(fig)

def dropdown_plan_eventhandler(change):
    common_filtering(change.new)

dropdown_planes_1.observe(dropdown_plan_eventhandler, names='value')

display(dropdown_planes_1)
display(plot_output)
display(output)
```
